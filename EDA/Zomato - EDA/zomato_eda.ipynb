```python
ratings = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings.head()
```

## Observation:
1. When Rating is between 4.6 - 4.9   ---->   Excellent
2. When Rating is between 4.0 - 4.5   ---->   Very Good
3. When Rating is between 3.5 - 3.9   ---->   Good
4. When Rating is between 2.5 - 3.4   ---->   Average
5. When Rating is between 1.0 - 2.4   ---->   Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color', palette=['black', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```

```python
 # Count Plot
 sns.countplot(x='Rating color', data=ratings, palette=['black', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```
