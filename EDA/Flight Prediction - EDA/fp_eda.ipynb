### Categorical Features

Airline, Source, Destination, Additional_Info

The Problem with LabelEncoder is "False Hierarchy"
When we use le.fit_transform on Airline, Source, and Destination, we are telling the computer:

IndiGo = 1
Jet Airways = 2
Air India = 3

A Machine Learning model (like Linear Regression) will look at those numbers and think: "Hmm, Air India (3) is three times 'greater' than IndiGo (1)." It might try to find a mathematical relationship between the names of cities or airlines where none actually exists.

```python
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
```

```python
# df['Airline'] = le.fit_transform(df['Airline'])
# df['Source'] = le.fit_transform(df['Source'])
# df['Destination'] = le.fit_transform(df['Destination'])
# df['Additional_Info'] = le.fit_transform(df['Additional_Info'])
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

cols = ['Airline', 'Source', 'Destination', 'Additional_Info']

encoded_array = ohe.fit_transform(df[cols])

encoded_df = pd.DataFrame(encoded_array, columns=ohe.get_feature_names_out(cols))

df = pd.concat([df.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)
df.drop(columns=cols, inplace=True)
```
